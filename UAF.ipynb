```python
import mysql.connector
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import spacy

# Download the English language model for spaCy
# python -m spacy download en_core_web_sm

# Load the English NLP model for spaCy
nlp = spacy.load("en_core_web_sm")

# Establishing the database connection
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Hari@25082002",
    database="library_database"
)

# Creating a cursor object
cursor = mydb.cursor()

# SQL query to fetch all subjects from the 'books_uaf' table
query = "SELECT subject FROM books_uaf"

# Executing the query
cursor.execute(query)

# Fetching all rows from the result set
rows = cursor.fetchall()

# Storing the subjects in a list
subjects = [row[0] for row in rows]

print(subjects)

# Getting user input
user_input = input("Enter a subject: ")
print("User Input:", user_input)

# Tokenizing the input
token = nltk.word_tokenize(user_input)
print("Tokenized Input:", token)

# Lemmatizing the words
lemma = WordNetLemmatizer()
lemmatized_words = [lemma.lemmatize(word) for word in token] 
print("Lemmatized Input:", lemmatized_words)

# Removing stop words
stop_words = set(stopwords.words('english')) 
filtered_words = [word for word in lemmatized_words if word.lower() not in stop_words]
print("Filtered Input:", filtered_words)

# Use spaCy for named entity recognition
doc = nlp(user_input)
entities = [entity.text for entity in doc.ents]
print("Named Entities:", entities)

# Checking if the user input matches any subject from the list
matched_subjects = []
for subject in subjects:
    if all(word.lower() in subject.lower() for word in filtered_words) or any(entity.lower() in subject.lower() for entity in entities):
        matched_subjects.append(subject)

if matched_subjects:
    print("Found subjects:", matched_subjects)
else:
    print("No matching subjects found.")
```
